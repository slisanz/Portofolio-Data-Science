# 02 Cleaning + geocoding

Apply the cleaning pipeline and write a parquet that the rest of the project reads.

```python
import sys, os
from pathlib import Path
ROOT = Path.cwd().parent
sys.path.insert(0, str(ROOT / 'src'))
%load_ext autoreload
%autoreload 2
```

```python
from ftth_equity import config, io, cleaning, geo
config.ensure_dirs()
raw = io.load_raw()
clean = cleaning.clean(raw)
clean = geo.to_wgs84(clean)
print(clean.shape, clean['is_lagging'].mean())
clean.head()
```

```python
out = config.INTERIM / 'buildings_clean.parquet'
io.save_parquet(clean, out)
out
```
