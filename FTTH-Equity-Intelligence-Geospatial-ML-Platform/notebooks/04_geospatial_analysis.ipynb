# 04 Geospatial analysis

Hex aggregation with H3 and a quick spatial-autocorrelation check on the lag rate.

```python
import sys, os
from pathlib import Path
ROOT = Path.cwd().parent
sys.path.insert(0, str(ROOT / 'src'))
%load_ext autoreload
%autoreload 2
```

```python
from ftth_equity import config, io, geo
df = io.load_parquet(config.INTERIM / 'buildings_clean.parquet')
```

## H3 cells


```python
# requires h3 v3
from ftth_equity.geo import add_h3
df = add_h3(df, resolution=9)
hex_stats = df.groupby('h3').agg(n=('imb_id','count'), lag=('is_lagging','mean'))
hex_stats.head()
```
